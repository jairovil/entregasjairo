```python
lista_tarefas = []
def adicionar_tarefa(lista_tarefas, descricao, status, prioridade):
    if not lista_tarefas:
        novo_id = 1
    else:
        novo_id = max(tarefa['id'] for tarefa in lista_tarefas) + 1

    nova_tarefa = {
        "id": novo_id,
        "descricao": descricao,
        "status": status,
        "prioridade": prioridade
    }

    lista_tarefas.append(nova_tarefa)

def visualizar_tarefas(lista_tarefas):
    for tarefa in lista_tarefas:
        print(f"ID: {tarefa['id']}, Descrição: {tarefa['descricao']}, Status: {tarefa['status']}, Prioridade: {tarefa['prioridade']}")


def filtrar_tarefas(lista_tarefas, status=None, prioridade=None):
    tarefas_filtradas = lista_tarefas
    if status:
        tarefas_filtradas = [tarefa for tarefa in tarefas_filtradas if tarefa['status'] == status]
    if prioridade:
        tarefas_filtradas = [tarefa for tarefa in tarefas_filtradas if tarefa['prioridade'] == prioridade]

    return tarefas_filtradas

if __name__ == "__main__":
 

    adicionar_tarefa(lista_tarefas, "Desenvolver App", "pendente", "alta")
    adicionar_tarefa(lista_tarefas, "Criar interface do usuário", "em andamento", "média")
    adicionar_tarefa(lista_tarefas, "Testar integração", "pendente", "baixa")
    adicionar_tarefa(lista_tarefas, "teste no ambiente de produção", "concluída", "alta")
    

    print("Todas as Tarefas:")
    visualizar_tarefas(lista_tarefas)
    print()

    print("Tarefas Pendentes com Alta Prioridade:")
    tarefas_filtradas = filtrar_tarefas(lista_tarefas, status="pendente", prioridade="alta")
    visualizar_tarefas(tarefas_filtradas)
    print()


    print("Tarefas Em Andamento:")
    tarefas_filtradas = filtrar_tarefas(lista_tarefas, status="em andamento")
    visualizar_tarefas(tarefas_filtradas)
    print()

    print("Tarefas com Baixa Prioridade:")
    tarefas_filtradas = filtrar_tarefas(lista_tarefas, prioridade="baixa")
    visualizar_tarefas(tarefas_filtradas)
```
